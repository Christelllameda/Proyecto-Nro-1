## Observamos que en los tipos de ataque se repite boating pero hay 2 elementos mal escritos, por lo que hacemos una funci√≥n para escribirlos bien y tener las categor√≠as bien

## Especies

```python
def limpiar_species(x):
    
    x = x.lower()
    
    if 'white shark' in x:
        return 'white shark'
    
    elif 'tiger shark' in x:
        return 'tiger shark'
    
    elif 'bull shark' in x:
        return 'bull shark'
    
    elif 'lemon shark' in x:
        return 'lemon shark'
    
    elif 'nurse shark' in x:
        return 'nurse shark'
    
    elif 'mako shark' in x:
        return 'mako shark'

    elif 'reef shark' in x:
        return 'reef shark'
    else:
        return x
```

```python
data['Species '] = data['Species '].apply(limpiar_species)
data['Species '].unique()
```

```python
def limpiar_fecha(x):
    
    x = x.lower()
    
    if 'm' in x:
        return 'M'
    elif 'f' in x:
        return 'F'
    if 'n' in x:
        return 'N'
    elif 'f' in x:
        return 'F'
    else:
        return x
```

```python
data['Sex '] = data['Sex '].apply(limpiar_fecha)
data['Sex '].unique()
```

```python
def try_parsing_date(text):
    for fmt in ('%d-%b-%Y', '%b %Y'):
        try:
            return pd.to_datetime(parse(text).strftime('%d-%b-%Y'), format='%d-%b-%Y')
        except ValueError:
            pass
    return pd.NaT  # Valor nulo si no se puede parsear

data['Date'] = data['Date'].apply(try_parsing_date)
```
